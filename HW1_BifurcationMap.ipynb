```python
import numpy as np
import matplotlib.pyplot as plt
```

1. Implement the map, plot the evolution of x

```python
n = 200
r = 2.6
x = np.zeros(n)
x[0] = 0.4

def find_x(x, n, r):
    for i in range(1, n):
        x[i] = r * x[i-1] * (1 - x[i-1])
    return x

x = find_x(x, n, r)
plt.figure(figsize=(12,6))
plt.title('Evolution of X with r=2.6')
plt.plot(x)
plt.xlabel('Iteration')
plt.ylabel('X')
plt.grid()
plt.show()
```

2. Create a linspace of r's, for every r save the last "m" values of x after the first "n" values (can be m=200, n=200), play around with values

```python
m = 200
n = 200
r = np.linspace(0, 4, 100)
x = np.zeros((len(r), m))
def find_x(n, m, r):
    x = np.zeros(n+m)
    x[0] = np.random.rand()
    for i in range(1, n + m):
        x[i] = r * x[i-1] * (1 - x[i-1])
    x_saved = x[n:]
    return x_saved

for i in range(len(r)):
    x[i] = find_x(n, m, r[i])


plt.figure(figsize=(12,6))
plt.title(f'Evolution of X with r={r[70]}')
plt.plot(x[70, :])
plt.xlabel('iterations')
plt.ylabel('X')
plt.grid()
plt.show()

plt.figure(figsize=(12,6))
plt.title(f'Evolution of X with r={r[80]}')
plt.plot(x[80, :])
plt.xlabel('iterations')
plt.ylabel('X')
plt.grid()
plt.show()

plt.figure(figsize=(12,6))
plt.title(f'Evolution of X with r={r[95]}')
plt.plot(x[95, :])
plt.xlabel('iterations')
plt.ylabel('X')
plt.grid()
plt.show()
```

3. Plot the bifurcation map

```python
plt.figure(figsize=(12,6))
plt.title('Bifurcation map')
plt.plot(r[:], x[:, :], 'r.', markersize=1)
plt.xlabel('r')
plt.ylabel('X')
plt.grid()
plt.show()
```
